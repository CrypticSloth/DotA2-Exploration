```python
import pandas as pd
import numpy as np
import json
import os
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from sklearn.preprocessing import OneHotEncoder
from mlxtend.preprocessing import OnehotTransactions
```

```python
# Load in the data
with open('Assoc_Rule_DataF_707d.json','r') as file:
        data = json.load(file)
```

```python
test = data
test[0]
```

```python
list = []
tempList = []
# Coding is: Order_Pick(T/F)_Hero_Team(0/1)
for i in range(len(test)):
    for x in range(len(test[i][0])):
        tempList.append('{}{}{}{}'.format(test[i][0][x]['order'],test[i][0][x]['is_pick'],test[i][0][x]['hero_name'],test[i][0][x]['team']))
        if test[i][1] == True:
            tempList.append('Rad_Win')
        else:
            tempList.append('Dire_Win')
            
    list.append(tempList)
    tempList = []
```

Do One Hot Encoding on the data so that we can run the apriori algorithm

```python
oht = OnehotTransactions()
oht_ary = oht.fit(list).transform(list)
df = pd.DataFrame(oht_ary, columns=oht.columns_)
```

Because there are a lot of different heros and draft pick orders all of the observations will have quite low support. However, we can still see which ones occur the most often.

```python
apriori(df,min_support=.05,use_colnames = True,max_len = 2)
```

What occurs the most often?

```python
frequent_itemsets = apriori(df, min_support=0.03, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

Lets filter our results to show some associations between features by setting 'length' to 2 and sorting the results.
Now we can see if certain picks at certain times during the draft cause teams to win more often.

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) ].sort_values('support',ascending=False)
```

It looks like the top result is that Radiant wins 4.7% of the time when they ban Night Stalker first. This can make sense as this data comes from pub matches and during patch 7.07d Night Stalker was quite strong in pub matches.
